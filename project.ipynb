4) Group the wines by "quality" and find the average value for "alcohol", "pH", "density" for every group (for both wines)

White wine

```python
%%sql
SELECT AVG(alcohol) as Average_alcohol, AVG(pH) as Average_pH, AVG(density) as Average_density, quality
from white_wine
GROUP BY "quality"
```

Red wine

5) Group one type of wine by "quality" and show only the groups with more than 50 wines

```python
%%sql
SELECT COUNT(*) as Total, quality
from white_wine
GROUP BY quality
HAVING Total > 50
ORDER BY Total
```

6) Create a CTE that groups by "quality" and calc. the avg for "acohol". Use the CTE to show only alcohol avg > 11

```python
%%sql
WITH avg_alcohol as (
    SELECT AVG(alcohol) as Average_alcohol, quality
    from white_wine
    GROUP BY quality
)
SELECT *
FROM avg_alcohol
WHERE Average_alcohol > 11
```

7) Create a column where we group wines by their quality. If we have a quality >= 7 we have a "High Quality", if the quality is <7 we have a "Bad Quality"

```python
%%sql
SELECT 
CASE 
WHEN quality >=7 THEN "Good Quality"
ELSE "Bad Quality"
END as Quality_report,
COUNT(*) as Total_wines
from white_wine
GROUP BY Quality_report
```

- We can make this extra specific and add more cases

```python
%%sql
SELECT 
CASE 
WHEN quality >= 8 AND alcohol >=12 THEN "Excelent wine"
WHEN quality >=6 AND quality <8 AND alcohol >10 THEN "Verry good wine"
ELSE "Bad wine"
END as Wine_clasification,
COUNT(*) as total_wines
from white_wine
GROUP BY Wine_clasification
ORDER BY total_wines
```

8) Window function - Add a column that shows for each quality the avg level of alcohol

```python
%%sql
SELECT quality, alcohol, AVG(alcohol) OVER(PARTITION BY quality) as Average_alcohol_for_each_quality
from white_wine
LIMIT 30 
```

9) Create a new csv file with some extra data, use join to get some data from both. We create a new csv file named "wine_types.csv". In this file we will have the following columns: quality(int), type(text: 'premium', 'standard', 'economy'). In the end show for each wine quality: the number of wines, avegare alcohol and the average volatile acidity.

 - First we read the new csv that I've created.

```python
df = pd.read_csv("wine_types.csv")
df.to_sql("wine_types", con, if_exists='replace', index=False)
```

```python
%%sql
SELECT * 
from wine_types
```

- It worked! Now we can do the joint

```python
%%sql
SELECT  WF.type, COUNT(*) as Total_number_of_wines, AVG(WW.alcohol), AVG(WW.[volatile acidity])
from white_wine as WW JOIN wine_types as WF ON WW.quality = WF.quality 
GROUP BY WF.type
```

11) For each quality show the first 5 wines with the highest alcohol

```python
%%sql
WITH RankedWines as (
  SELECT quality, alcohol,
  ROW_NUMBER() OVER(PARTITION BY quality ORDER BY alcohol desc) as rank
  from white_wine
)
SELECT * from RankedWines WHERE rank <=5
```

12) Calculate the cumulative sum for alcohol, order by wine quality

```python
%%sql
SELECT quality, alcohol,
SUM(alcohol) OVER (ORDER BY quality) as cumulative_sum_alc
from white_wine LIMIT 30
```
